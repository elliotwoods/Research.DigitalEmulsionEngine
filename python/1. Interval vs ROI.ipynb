# logInterval

##Abstract

The Ximea xiQ device can consistently capture 2048x1536 frames at 120Hz using the driver for MS Windows (in free run mode), performance is reduced in OSX. The frame interval (time taken to procure a frame) is found to be proportional to the ROI (Region of Interest) size. ROI can only be affected in the Y direction (i.e. Y offset, height).

##Introduction

The code loops different sizes of ROI and different exposures. Sometimes the program crashes, but can be restarted with a  different `roiY` starting value in order to resume from where it left off

## Import data

```python
import glob
filenames = glob.glob("../c++/logInterval/data-win32/*.txt")
# filenames = glob.glob("../c++/logInterval/data/*.txt")

for filename in filenames:
    A = loadtxt(filename)
    A = A[10:,:] # strip first 10 datapoints as camera initialises
    
    if filename == filenames[0]:
        dataSet = A
    else:
        dataSet = vstack([dataSet,A])
dataSet.shape
backup = dataSet
print dataSet.shape
```

```python
# Recover dataSet
dataSet = backup
dataSet.shape
```

Data arrives in the format:

* 0 - availableBandwidth
* 1 - limitBandwidth
* 2 - exposure
* 3 - roiY
* 4 - roiHeight
* 5 - nFrame
* 6 - interval (system)
* 7 - intervalTimestamp

##Clean out erroneous points

Sometimes frames arrive corrupted, and since the frame includes a timestamp, this also results in an erroneous interval. We need to ignore these frames from our dataset (which can be identified by a late

```python
# strip where timestamp interval is above 1s
dataSetStripped = delete(dataSet, where(dataSet[:,7] > 1e9), 0)
print dataSetStripped.shape
```

## Optional : Remove low ROI points

Useful for fitting data without all the outliers at the start of the series

```python
# strip where roi is below 200
dataSetStripped = delete(dataSetStripped, where(dataSetStripped[:,4] < 200), 0)
print dataSetStripped.shape
```

## Plot best interval vs ROI

Include a 1st order polyfit line.

```python
# zoom to bottom left
# xlim([0, 1600])
# ylim([0, 50000])

rois = unique(transpose(dataSetStripped[:,4]))
print rois.shape

for roi in rois:
    first = (roi == rois[0])
    selectROI = where(dataSetStripped[:,4] == roi)
    intervals = transpose(dataSetStripped[selectROI, 7] / 1000)
    lowIntervals = sort(intervals)[0:40]
    count = lowIntervals.shape[0]
    thisx = np.ones(count) * roi
    thisy = transpose(lowIntervals)
    if first:
        x = thisx
        y = thisy
    else:
        x = hstack([x, thisx])
        y = hstack([y, thisy])
x = hstack(x)
y = hstack(y)

plot(x,y, '.')
title("Best intervals vs ROI")
xlabel("ROI y (px)")
ylabel("Interval (us)")

# find lowest interval and mark it
minInterval = min(y);
minIntervalLocation = x[where(y == minInterval)[0][0]]
print "Minimum interval is " + str(minInterval) + "us, which gives an fps of " + str(1e6 / minInterval) + "Hz"
plot(minIntervalLocation, minInterval, 'o', label="Minimum interval", markersize=20, markerfacecolor='white')

# polyfit
fit = polyfit(hstack(x), hstack(y), 1)
polyfn = poly1d(fit)
xfitline = linspace(xmin, xmax, 100)
plot(xfitline, polyfn(xfitline), label="Fit")

# show 120Hz line
xmin, xmax = xlim()
targetInterval = 1e6 / 120
plot([xmin, xmax], [targetInterval, targetInterval], label="120Hz")

# show 1536px line
ymin, ymax = ylim()
targetROI = 1536
plot([targetROI, targetROI], [ymin, ymax], label="1536px")

legend()
grid(True)
      
print "We estimate:"
print "1. It takes " + str(fit[1]) + "us per line of pixels"
print "2. There is an overhead of " + str(fit[0]) + "us for eash frame read out"
print "3. Use `estimateFps(roiHeight)` to give an fps for an roi"

def estimateFps(roiHeight):
    return 1e6 / polyfn(roiHeight)
```
