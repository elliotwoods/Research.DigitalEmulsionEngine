# 3. Colour wheel cycle timings

## Introduction

### Colour wheels
Economy DLP projectors use a single light path to present colour images (and therefore have only 1 DLP chip). In order to present colour images, for each frame they present a few quick subframes, each of which is a different colour. There is 1 roration of the colour wheel per projected frame. Information can be found at http://en.wikipedia.org/wiki/Digital_Light_Processing#Single-chip_projectors

## Experiment
In this experiment we aim to recover the relationship between input colour colour data and output per colour wheel channel (on our projector, these channels are red, green, blue, white, magenta, cyan).

### Framestamp
First we generate a colour image with a frame index using application showFrameIndex. On the projector this should increase at 120fps (here on a monitor it increases at 60fps). We can use the frame index as a reference for which application frame is being presented. With vertical sync enabled, application frame index should equate to projector frame index if render time is sufficiently small, but this is not guaranteed here.

### Load data and define a plot function

We also draw the raw data as a test

````python
filename = dataPath + "120Hz-1us-2s.txt"
dataSet = loadtxt(filename)

#time is in microseconds currently
timeMagnitude = 1e-6
dataSet[:,0] *= timeMagnitude

def setupFigure():
    fig = figure()
    axes = fig.add_subplot(1, 1, 1, axisbg='black')
    return axes
    
def plotSeries():
    plot(dataSet[:,0],dataSet[:,1], 'r-', label="Red")
    plot(dataSet[:,0],dataSet[:,2], 'g-', label="Green")
    plot(dataSet[:,0],dataSet[:,3], 'b-', label="Blue")
    plot(dataSet[:,0],dataSet[:,4], 'y-', label="Yellow")
    plot(dataSet[:,0],dataSet[:,5], 'c-', label="Cyan")
    plot(dataSet[:,0],dataSet[:,6], 'm-', label="Magenta")
    plot(dataSet[:,0],dataSet[:,7], 'w-', label="White")

def finaliseFigure(xaxis='time (s)'):
    ylim([0, 255])
    ylabel('brightness (/255)')
    xlabel(xaxis)
    grid(True, which='both', axis='both', color='w')
    legend(loc='upper center')

def plotDataSet(xaxis='time (s)'):
    setupFigure()
    plotSeries()
    finaliseFigure(xaxis)
    
plotDataSet()
````

### Stack frames at interval (e.g. 120Hz)

And for convenience, we convert the time value into cycles (presuming constant angular velocity).

````python
projectorFrameRate = 120
wheelInterval = 1.0 / projectorFrameRate

for row in range(dataSet.shape[0]):
    timestamp = dataSet[row,0]
    nPeriods = floor_divide(timestamp, wheelInterval)
    dataSet[row,0] -= nPeriods * wheelInterval
    
sortedIndices = argsort(dataSet[:,0])
dataSet = dataSet[sortedIndices,:]
dataSet[:,0] /= wheelInterval

plotDataSet(xaxis='cycles')
````

Here we can see the output assignment as time in x, given an input value as colour in series. For instance, we note that when presenting white (the white series), we use many parts of the colour cycle, whilst when we present blue, we use only one part of the cycle (which we presume to be when the blue gel ordinarily interrupts the light path).

We first note the order of primary colours to recover the rotation direction of the wheel.

````python
setupFigure()
plot(dataSet[:,0],dataSet[:,1], 'r.', label="Red")
plot(dataSet[:,0],dataSet[:,2], 'g.', label="Green")
plot(dataSet[:,0],dataSet[:,3], 'b.', label="Blue")
finaliseFigure()
print "Mean brightness per cycle in red = " + str(mean(dataSet[:,1]))
````

Noting that the IR registration marker on the wheel goes high between blue and white and continues into Blue, we describe the order as:

The order of the colours on the wheel can then be presumed as:
1. Blue
2. Red
3. Yellow
4. Green
5. Magenta
6. White

I.e. a camera sync signal taken from the registration mark reader would cause the camera to trigger on Blue. This is advantageous since blue's colour is presented for the shortest duration of time (and therefore can be least visible).

### Possible issue?

If we were to attempt to send the signal on a single colour (e.g. blue) and the anti-signal on another colour (e.g. red+green), then we'd essentially be presenting a yellow or magenta, which the projector would then interpret as a mix of colours, e.g.:

* Present is 230 for this pixel
* Spatial encoding is low for this pixel
* Blue is 0
* Red + Green need to counteract, so present 230 / 2 = 115 each
* Colour is processed by projector as Yellow giving the following response (a mix of red, green and yellow):

````python
setupFigure()
plot(dataSet[:,0],dataSet[:,4], 'y.', label="Yellow")
finaliseFigure()
print "Mean brightness per cycle in yellow = " + str(mean(dataSet[:,4]))
````

````python
setupFigure()
plot(dataSet[:,0],dataSet[:,5], 'c.', label="Cyan")
finaliseFigure()
print "Mean brightness per cycle in cyan = " + str(mean(dataSet[:,5]))
````

This is great! We can send data using the following truth table

```
            Image high     Image low
Data high   Cyan           Low blue*
Data low    Yellow         Low red
```

* Lowest that can be perceived by camera

We can then consider the visual correction as seperate from the balanced graycode signal, i.e. 2 frames are still necessary for a signal + and signal - pair, but both types of signal can be imperceptible to the viewer, thereby allowing for 120Hz content to be presented.

Functions then are fitted as:
1. Input value -> position on cyan to low blue scale
2. Position on cyan to low blue -> Position on yellow to red

i.e. one of these ranges defines our available dynamic range, and the other should be matched to that, so that for every output pixel we can select where the brightness will be presented (i.e. including blue or excluding blue).

Cyan offers a considerable increase in brightness over blue (31 vs 18), which means that selecting it increases the upper limit of our dynamic range for the output.

### Estimating the period of blue

We'll be syncing to blue (as it is has the closest connection to the IR sync signal). And our anti-signal will be presented in yellow which is presented immediately after blue. Therefore we must acquire an accurate exposure duration for blue to be captured for. We presume this generally to be performed with a runtime algoritem which optimises:

1. Phase between sync input from projector to sync output to camera
2. Duration of exposure

````python
# A time series where red and blue are centred
dataSetBlueCentered = copy(dataSet[:,0])
dataSetBlueCentered += 0.5

for i in arange(dataSetBlueCentered.size):
    if dataSetBlueCentered[i] > 1:
        dataSetBlueCentered[i] -= 1

setupFigure()
plot(dataSetBlueCentered,dataSet[:,1], 'r.', label="Red")
plot(dataSetBlueCentered,dataSet[:,3], 'b.', label="Blue")

plot([0.345, 0.5], [80, 80], 'w-')

finaliseFigure()
print "Mean brightness per cycle in red = " + str(mean(dataSet[:,1]))

print string(((0.5 - 0.345) * wheelInterval) * 1e6) + "us"
````
