# 3. Colour wheel cycle timings

## Introduction

### Colour wheels
Economy DLP projectors use a single light path to present colour images (and therefore have only 1 DLP chip). In order to present colour images, for each frame they present a few quick subframes, each of which is a different colour. There is 1 roration of the colour wheel per projected frame. Information can be found at http://en.wikipedia.org/wiki/Digital_Light_Processing#Single-chip_projectors

## Experiment
In this experiment we aim to recover the relationship between input colour colour data and output per colour wheel channel (on our projector, these channels are red, green, blue, white, magenta, cyan).

### Framestamp
First we generate a colour image with a frame index using application showFrameIndex. On the projector this should increase at 120fps (here on a monitor it increases at 60fps). We can use the frame index as a reference for which application frame is being presented. With vertical sync enabled, application frame index should equate to projector frame index if render time is sufficiently small, but this is not guaranteed here.

If we want to observe the time multiplexing of colour values, we must capture at a higher sample rate than the rate of the projector, e.g.:

* Camera at 360Hz = 3 captures per rotation
* Camera at 600Hz = 5 captures per rotation
* Camera at 1200Hz = 10 captures per rotation

We need sufficient captures per rotation to reslove the patterns for each of the 6 channels in our colour wheel (see Experiment 2).

### Load data and define a plot function

We also draw the raw data as a test

```python
filename = dataPath + "120Hz.txt"
dataSet = loadtxt(filename)

#time is in microseconds currently
timeMagnitude = 1e-6
dataSet[:,0] *= timeMagnitude

def setupFigure():
    fig = figure()
    axes = fig.add_subplot(1, 1, 1, axisbg='black')
    
def plotSeries():
    plot(dataSet[:,0],dataSet[:,1], 'r-', label="Red")
    plot(dataSet[:,0],dataSet[:,2], 'g-', label="Green")
    plot(dataSet[:,0],dataSet[:,3], 'b-', label="Blue")
    plot(dataSet[:,0],dataSet[:,4], 'y-', label="Yellow")
    plot(dataSet[:,0],dataSet[:,5], 'c-', label="Cyan")
    plot(dataSet[:,0],dataSet[:,6], 'm-', label="Magenta")
    plot(dataSet[:,0],dataSet[:,7], 'w-', label="White")

def finaliseFigure():
    ylim([0, 255])
    ylabel('brightness (/255)')
    xlabel('time (s)')
    grid(True, which='both', axis='both', color='w')
    legend(loc='upper center')

def plotDataSet():
    setupFigure()
    plotSeries()
    finaliseFigure()
    
plotDataSet()
```

### Stack frames at interval (e.g. 120Hz)

```python
projectorFrameRate = 120
interval = 1.0 / projectorFrameRate

for row in range(dataSet.shape[0]):
    timestamp = dataSet[row,0]
    nPeriods = floor_divide(timestamp, interval)
    dataSet[row,0] -= nPeriods * interval
    
sortedIndices = argsort(dataSet[:,0])
dataSet = dataSet[sortedIndices,:]

plotDataSet()
```

Here we can see the output assignment as time in x, given an input value as colour in series. For instance, we note that when presenting white (the white series), we use many parts of the colour cycle, whilst when we present blue, we use only one part of the cycle (which we presume to be when the blue gel ordinarily interrupts the light path).

```python
setupFigure()
plot(dataSet[:,0],dataSet[:,3], 'b-', label="Blue")
finaliseFigure()
```

We experience significant ramping on this data, perhaps this is due to shutter speed? Let's calculate if this could be shutter speed.

```python
interval = 1.0 / 120
maxShutter = interval

observedRamp = 0.001

print [maxShutter, observedRamp, observedRamp/maxShutter]
```

Yes, since the observed ramp is a sub-portion of the possible shutter speed, it's highly likely that the shutter speed is regularly straddling binary values (hence being observed as a linear ramp). Side note: since the ramp is almost linear, we can presume that our camera response is also almost linear (photons captured ~ pixel value).
