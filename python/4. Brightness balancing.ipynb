# 4. Balance brightness

The purpose of this experiment is to find a transfer function `f(A)` where:

```
p([R,G,B]) = pereceived intensity of projected colour x
p([f(A), f(A), 0]) == p([0, 0, A])
```

This transfer function therefore allows us to cancel signals sent in the blue channel via signals sent in the R,G channels. Also we note:

```
p([0,0,A]) = Black level of projection image (lowest possible brightness)
255 - f(A) = Available colour values for sending image data
```

````python
dataSet = loadtxt(dataPath + "optoma-brightness rg,b.txt")

# observed
plot(dataSet[:,0],dataSet[:,1], '.', label="observed")

# fit
fit = polyfit(dataSet[:,0],dataSet[:,1],1)

# axes
xlabel("Signal value A")
ylabel("Balancing value f(A)")

grid(True)
legend()

print fit
````
